```python
import re

def clean_plot(text):
    text = text.lower()
    text = re.sub(r"\[\d+\]", "", text)  # remove [1], [2]
    text = re.sub(r"\{\{.*?\}\}", "", text)  # templates
    text = re.sub(r"==.*?==", "", text)  # headings
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
df1["clean_plot"] = df1["Plot"].apply(clean_plot)
```

```python
df1["Director"] = df1["Director"].fillna("").str.strip()
df1["Cast"] = df1["Cast"].fillna("").str.strip()
df1["Genre"] = df1["Genre"].fillna("").str.strip()
    
```

```python
df1.isna().sum()
```

```python
def normalize_cast(cast):
    if not cast or cast.strip() == "":
        return []
    
    actors = cast.split(",")
    actors = [a.strip() for a in actors if a.strip()]
    return actors

```

```python
def normalize_genre(genre):
    if not genre or genre.strip() == "":
        return []
    
    genres = genre.split(",")
    genres = [g.strip().title() for g in genres if g.strip()]
    return genres
```

```python
df1["cast_list"] = df1["Cast"].apply(normalize_cast)
df1["genre_list"] = df1["Genre"].apply(normalize_genre)
```
